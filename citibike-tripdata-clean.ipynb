```python
# Import Dependencies
import pandas as pd
from pathlib import Path

# Import citi bike CSV files
jc_202301_CBtripdata = Path("Resources/JC-202301-citibike-tripdata.csv")
jc_202307_CBtripdata = Path("Resources/JC-202307-citibike-tripdata.csv")

# Read in citi bike trip data and store in Pandas DataFrame
jan_tripdata_df = pd.read_csv(jc_202301_CBtripdata)
jul_tripdata_df = pd.read_csv(jc_202307_CBtripdata)
```

```python
# Print dataframe
jan_tripdata_df.head()
```

```python
# Print dataframe
jul_tripdata_df.head()
```

```python
# Drop the ride_id colmn from the dataframes because it is not needed
jan_tripdata_dfc = jan_tripdata_df.drop(columns=['ride_id', 'rideable_type'])
jan_tripdata_dfc = jan_tripdata_dfc.dropna()
jan_tripdata_dfc = jan_tripdata_dfc.set_index('start_station_id')
jan_tripdata_dfc.head()
```

```python
# Drop the ride_id colmn from the dataframes because it is not needed
jul_tripdata_dfc = jul_tripdata_df.drop(columns=['ride_id', 'rideable_type'])
jul_tripdata_dfc = jul_tripdata_dfc.dropna()
jul_tripdata_dfc = jul_tripdata_dfc.set_index('start_station_id')
jul_tripdata_dfc.head()
```

```python
# Combine the first two DataFrames into one single DataFrame
jan_jul_df = pd.merge(jan_tripdata_dfc, jul_tripdata_dfc, on='start_station_id', how='inner')
jan_jul_df.head()
```

```python
# Export the data frane into a CSV file to import to tableau
jan_jul_df.to_csv('combined_cbtripdata.csv', index=True)
```
